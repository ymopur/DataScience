# Analysis of Nutrients

### Extracting the data files using Pandas

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

### Ensuring that the encoding is set to enable proper import of the file

```python
agri=pd.read_csv('ABBREV.csv', encoding = "ISO-8859-1", index_col='NDB_No')
```

### Checking the first 5 rows of the dataframe

```python
agri.head()
```

### Searching for a particular character using 'str' function in Pandas and entering relevant values into a new column 'Group'

```python
agri['Group'][agri['Shrt_Desc'].str.contains("CHEESE")]='Cheese'
```

```python
agri['Group'][agri['Shrt_Desc'].str.contains("YOGURT")]='Yogurt'
```

```python
agri['Group'][agri['Shrt_Desc'].str.contains("BUTTER")]='Butter'
```

```python
agri['Group'][agri['Shrt_Desc'].str.contains("MILK")]='Milk'
```

### Creating a new dataframe containing only the relevant groups

```python
agri1=agri[agri['Group'].isin(['Milk','Butter','Yogurt','Cheese'])]
```

### Converting a normal dataframe to a groupby dataframe

```python
agri1=agri1.groupby('Group')
```

```python
agri1.head()
```

### Creating a new dataframe containing the same columns as the original dataframe

```python
df = pd.DataFrame(columns=agri1.columns)
df
```

### Extract the records containing highest protein content

```python
for food,protein in agri1:
    highest_protein=protein.nlargest(4,"Protein_(g)")
    df=df.append(highest_protein).sort_values('Group')
```
